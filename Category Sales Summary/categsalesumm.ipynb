```python
import pandas as pd
import numpy as np
```

```python
product = pd.read_csv("../products.csv")
product.head()
```

```python
sales = pd.read_csv("../sales.csv")
sales.head()
```

```python
merged = pd.merge(product,sales, on='product_id', how='inner')
merged.head()
```

```python
merged['revenue'] = merged['units_sold']* merged['price_per_unit']
merged.head()
```

```python
merged['average_price_per_unit'] = merged['revenue']/ merged['units_sold']

merged.head()
```

```python
merged.dtypes
```

```python
category_summary = merged.groupby('category').agg({
    'revenue': 'sum',
    'units_sold': 'sum',
    'average_price_per_unit' : 'mean'
})

category_summary.head()
```

```python
category_summary.to_csv("category_summary.csv", index= False)
```
